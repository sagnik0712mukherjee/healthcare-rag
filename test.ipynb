```python
class Solution:
    def reverse(self, x: int) -> int:

        lower = -1 * (2 ** 31)
        upper = (2**31) - 1
        
        str_int = str(x)
        initiator = ""
        if str_int[0] == '-':
            initiator = str_int[0]
            str_int = str_int[1:]
        
        rev_int = int(initiator + str_int[::-1])

        if rev_int < lower or rev_int > upper:
            return 0

        return int(rev_int)

sol = Solution()
sol.reverse(1534236469)
```
